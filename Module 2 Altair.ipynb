```python
import pandas as pd
data = pd.read_csv("CharlotteData.csv")
data.head()
```

```python
import altair as alt


base = alt.Chart(data).transform_window(order = 'count()').encode(alt.X('Period', sort = None))
bar = base.mark_bar(color = "pink").encode(alt.X('Period', sort = None), y = "Temperature", tooltip = ["Temperature"])


line = base.mark_line(color = "black", point = True).encode(y = "Relative Humidity",
                                                           tooltip = ["Relative Humidity"])
(bar + line).properties(width = 600)
```
